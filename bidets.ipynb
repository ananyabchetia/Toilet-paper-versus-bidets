```python
import pandas as pd

df=pd.read_csv('permits.csv')
df 
```

```python
contains_years = df['Issued Date'].astype(str).str.contains('2016', case=False, na=False).any()

print(contains_years)
```

## Gonna merge dataset to find zipcode

```python
import pandas as pd

df=pd.read_csv('pluto_25v1_1.csv')
df
```

## Make borough name values all capital letters. 

```python
borough_mapping = {
    'BX': 'BRONX',
    'BK': 'BROOKLYN',
    'MN': 'MANHATTAN',
    'QN': 'QUEENS',
    'SI': 'STATEN ISLAND'
}


df['borough'] = df['borough'].replace(borough_mapping)


df
```

```python
df.rename(columns={'borough': 'Borough'}, inplace=True)
df.rename(columns={'block': 'Block'}, inplace=True)
df.rename(columns={'lot': 'LOT'}, inplace=True)
```

```python
df
```

## Yay. time to merge to find the right zipcode. 

```python
import pandas as pd
df_1=pd.read_csv('rows_with_bidet.csv')
df_1
```

```python
merged_df = df_1.merge(df[['Borough', 'Block', 'LOT', 'zipcode']], 
                      on=['Borough', 'Block', 'LOT'], 
                      how='left')
merged_df
```

```python
merged_df.to_csv('bidet_demographics.csv')
```
